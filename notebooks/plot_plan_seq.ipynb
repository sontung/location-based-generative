```python
import matplotlib.pyplot as plt
from PIL import Image
plt.figure(figsize=(15, 15))


plt.subplot(3, 1, 1)
plt.axis("off")
plt.imshow(Image.open("s1.png").convert('RGB'))

plt.subplot(3, 1, 2)
plt.axis("off")
plt.imshow(Image.open("s2.png").convert('RGB'))

plt.subplot(3, 1, 3)
plt.axis("off")
plt.imshow(Image.open("s3.png").convert('RGB'))
plt.tight_layout()
plt.savefig("plan_bynlp.png", transparent=True, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.figure(figsize = (15,15))
gs1 = gridspec.GridSpec(3, 1)
gs1.update(wspace=0.025, hspace=0.05) # set the spacing between axes. 

for i in range(3):
   # i = i + 1 # grid spec indexes from 0
    ax1 = plt.subplot(gs1[i])
    plt.axis("off")
    plt.imshow(Image.open("s1.png").convert('RGB'))
    plt.title("['yellow', 'up', 'gray']")
    ax1.set_xticklabels([])
    ax1.set_yticklabels([])
    ax1.set_aspect('equal')

plt.show()
```
